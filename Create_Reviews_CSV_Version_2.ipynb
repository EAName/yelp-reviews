Review File

```python
review_json_path = '/content/gdrive/MyDrive/yelp_dataset/yelp_academic_dataset_review.json'
```

```python
data = {'user_id': [], 'business_id': [], 'date': [], 'stars': [], 'text': [], 'useful': [], 'funny': [],'cool': []}

with open(review_json_path) as f:
    for line in tqdm(f):
        review = json.loads(line)
        data['user_id'].append(review['user_id'])
        data['business_id'].append(review['business_id'])
        data['date'].append(review['date'])
        data['stars'].append(review['stars'])
        data['text'].append(review['text'])
        data['useful'].append(review['useful'])
        data['funny'].append(review['funny'])
        data['cool'].append(review['cool'])
```

```python
df_r = pd.DataFrame(data)

print(df_r.shape)
df_r.head()
```

```python
# only keep businesses that are categorized as restaurant
df_r = df_r.merge(df_b,left_on='business_id',right_on='business_id')
```

```python
df_r  = df_r.iloc[: , :8]
df_r = df_r.rename(columns = {'stars_x':'stars'})
```
